The dataset is only 8MB in size, so this is a fairly simple inspection.

```python
import numpy as np
import pandas as pd
```

```python
clothing_df = pd.read_csv(r'rough_data\anonymized_womens_clothing_reviews\womens_clothing_e-commerce_reviews.csv')
```

With the large positive skew exhibited by online user reviews, we need a significant number of reviews to increase the likelyhood of negative reviews being present at all. The negative reviews (1 and 2 out of 5) are likely to be only about 10% of the total reviews, according to prior research.

Now for some ratings info:

```python
clothing_df['Rating'].mean()
```

The overall average rating for clothing articles was 4.2

```python
clothing_df.groupby('Clothing ID')['Rating'].agg(['mean','count']).sort_values('count', ascending=False)[:50]
```

Looks like most articles of clothing with significant numers of reviews averaged ratings in the low to mid 4's.
