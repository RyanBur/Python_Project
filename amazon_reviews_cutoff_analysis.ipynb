```python
def sample_data_creation(num_ratings, num_samples, series_in=sample_item_df['rating']):
    '''
    Sample reviews from the population of reviews for a specific product.
    Default values are set for use with item 0439023153 (the intended use for the 
    function), but can be changed if needed.
    
    Inputs:
    num_ratings: Integer, the number of reviews in the output sample.
    num_samples: Integer, the number of samples in the output
    series_in: Input data
    '''
    output_list=[]
    for sample_number in range(num_samples): 
        review_subset = np.random.choice(series_in, num_ratings, replace=False)
        avg = review_subset.mean()
        output_list.append(avg)
    series_out = pd.Series(output_list)
    return series_out
```

```python
num_ratings = 25
sample_series = sample_data_creation(num_ratings,5000)
plt.hist(sample_series, bins=45, density=False, range=(3.3,4.8))
plt.title(f'Sample Average Rating Distribution\nFor Ratings/Sample of {num_ratings}', fontsize=16)
plt.xlabel('Sample Average Rating')
plt.ylabel('Frequency')

# Standard dev of the samples
std = np.std(sample_series)
# z-score for 90% confidence interval
z_score = 1.645
ci_range = z_score * std
print('90% Confident that the real product average lies within \n+-', \
      round(ci_range,2),f'from the observed average when the sample size is {num_ratings}')
```
